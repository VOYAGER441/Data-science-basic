```python
# 4x5 pizza grid: toppings coded as numbers (1=cheese, 2=pepperoni, 3=mushroom, 4=olive)
pizza = np.array([
    [1, 2, 3, 4, 1],  # Slice 0
    [2, 3, 1, 2, 3],  # Slice 1
    [4, 1, 2, 3, 4],  # Slice 2
    [3, 4, 1, 2, 1]   # Slice 3
])
```

```python
print("Full Pizza:\n", pizza)
```

```python
# Challenge 1: Serve First Two Slices
print(pizza[0:2])
print(pizza[:2,:])
```

```python
#  Cheese-Only Toppings (Column 0)
print(pizza[:, 0])
```

```python
# Challenge 3: Pepperoni Middles with Step
# Extract every other topping starting from column 1, but only from middle slices (rows 1:3).
print(pizza[1:3,1::2])
```

```python
#Challenge 4: Reverse a Slice
# Serve slice 2 backwards (reverse columns). 
 
print(pizza[2:3,::-1])
```
